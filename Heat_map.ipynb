# Gráficas de calor

El objetivo de una gráfica de calor es poder ubicar aquellos elementos que tengan 

```python
# Librerías necesarias para realizar una gráfia de calor
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Ejemplo usando números aleatorios

```python
# Matriz de números aleatorios, 4 filas por 6 columnas
data = np.random.rand(4, 6)
```

```python
print(data,"\n")
print(data.size)
```

```python
# Convertimos la matriz a un DataFrame
df=pd.DataFrame(data)
df
```

```python
df.shape
```

```python
# Asignamos nombres a las filas y columnas mediante el uso de 2 listas
indice=[1,2,3,4]
columnas=[1,2,3,4,5,6]

# Reescribimos el DataFrame, pero ahora usando las listas creadas para
# dar nombre a las filas y a las columnas:

df=pd.DataFrame(data,index=indice, columns=columnas)
df
```

```python
# Creamos el objeto heatmap, ahí se almacenará toda la información
# relacionada con la gráfica:

heat_map = sns.heatmap(df)
plt.show()
```

```python
heat_map = sns.heatmap(df, vmin=0, vmax=1)
plt.show()
```

```python
heat_map = sns.heatmap(df, vmin=0, vmax=1, cmap="YlGn")
plt.show()
```

### Ejemplo 3, Usando la base de datos del número de vuelos entre 1949 a 1960

```python
flights = pd.read_csv("Datasets/flights.csv")
flights.head()
```

```python
# Para poder manipular los datos, será necesario pivotear el DataFrame
# Al pivotear, le indicamos a pandas que una columna se use como index y
# otra como titulos de columnas

flights = flights.pivot("month", "year", "passengers")
flights
```

```python
flights.rename(index={"January":1, "February":2, "March": 3, "April":4, 
                     "May":5, "June":6, "July":7, "August":8, "September":9,
                     "October":10, "November":11, "December":12}, inplace=True)
flights.sort_index(inplace=True)
flights
```

```python
flights.rename(index={1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr", 
                     5:"May", 6:"Jun", 7:"Jul", 8:"Aug",
                      9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"}, inplace=True)
flights
```

```python
ax = sns.heatmap(flights, cmap="Reds")
plt.show()
```

De la gráfica anterior, podemos observar que los vuelos ocurren con mucha mayor frecuencia en los meses de Julio y Agosto. Además de que el número de vuelos a aumentado de manera general en todos los meses a lo largo de los años.

```python
# Agregamos los valores correspondientes a cada una de las celdas, para poder
# proporcionar más información sobre la gráfica:

ax = sns.heatmap(flights, annot=True, fmt="d", cmap="Reds")
plt.show()
```

```python
# Ajustamos el largo y el alto de la figura:
plt.figure(figsize=(15,8))
# Ajustamos el tamaño de letra:
sns.set(font_scale=1.3)
ax = sns.heatmap(flights, annot=True, fmt="d", cmap="Reds")
plt.show()
```
